## 6. Discount Calculation
Calculating discount amount and discount percentage for each product.


```python
df['discount'] = df['market_price'] - df['sale_price']
df['discount_percent'] = (df['discount'] / df['market_price']) * 100
df[['product','sale_price','market_price','discount_percent']].head(20)
                                                                
```

## 7. Missing Values

```python
df.isnull().sum()
```

## 8. Outlier Detection & Treatment
Using IQR method to detect extreme values and replacing them with column mean.


```python
numerical_cols = ['sale_price','market_price','rating']

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    mean_value = df[col].mean()
    df[col] = np.where((df[col] < lower) | (df[col] > upper), mean_value, df[col])

df.describe()
```

## 9. Visualizations
Creating simple plots to understand pricing and rating trends.


```python
plt.figure()
plt.hist(df['sale_price'], bins=30, edgecolor='black', color='LightGreen')
plt.title('Sale Price Distribution')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

```python
category_rating = df.groupby('category')['rating'].mean()

plt.figure()
category_rating.plot(kind='bar',edgecolor='black', color='Purple')
plt.title('Average Rating by Category')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.show()
```

## 10. Conclusion

- Discount strategy plays an important role in pricing.
